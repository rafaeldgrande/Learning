```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('presentation.mplstyle')
```

# Plotting lines and points

```python
x = np.array([1, 2, 3, 4, 5])
y = x + np.random.random(len(x))

y_error = np.random.random(len(x))/2
x_error = np.random.random(len(x))

pz_fit = np.polyfit(x, y, 1)
pz = np.poly1d(pz_fit)
```

```python
plt.figure()

plt.errorbar(x, y, yerr = y_error, xerr = x_error, fmt ='', elinewidth=2,
             capsize = 8, linestyle='')
plt.plot(x, y, 'ro', label='"data"')


plt.plot(x, pz(x), 'k--', label='fit', color='orange')

plt.xlabel(r'$x (\mathrm{\AA})$')
plt.ylabel(r'$y (\mathrm{eV/\AA})$')

for ix in range(len(x)):
    plt.axhline(y=y[ix], color='black', linewidth=0.6, alpha=0.6)
    plt.axvline(x=x[ix], color='black', linewidth=0.6, alpha=0.6)



plt.legend()
```
