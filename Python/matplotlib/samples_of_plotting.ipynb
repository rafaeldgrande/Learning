```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('presentation.mplstyle')
```

# Plotting lines and points

```python
# genereating some data with random variation around y=x
x = np.array([1, 2, 3, 4, 5])
y = x + np.random.random(len(x))

# error in x and y
y_error = np.random.random(len(x))/2
x_error = np.random.random(len(x))

# linear fit
pz_fit = np.polyfit(x, y, 1)
pz = np.poly1d(pz_fit)
```

```python
plt.figure()

# how to plot errorbars
# capsize is very important! Without it the errorbar is just a line
plt.errorbar(x, y, yerr = y_error, xerr = x_error, fmt ='', elinewidth=2,
             capsize = 8, linestyle='')
plt.plot(x, y, 'ro', label='"data"')

# dashed line
plt.plot(x, pz(x), 'k--', label='fit', color='orange')

plt.xlabel(r'$x (\mathrm{\AA})$')
plt.ylabel(r'$y (\mathrm{eV/\AA})$')

for ix in range(len(x)):
    plt.axhline(y=y[ix], color='black', linewidth=0.6, alpha=0.6)
    plt.axvline(x=x[ix], color='black', linewidth=0.6, alpha=0.6)



plt.legend()
```

```python
x = np.linspace(0, 2, 101)

plt.plot(x, x, '-', label='solid')
plt.plot(x, x**2, '--', label='dashed')
plt.plot(x, x**3, ':', label='pointed')

plt.legend()
```

# Plotting matrix

```python
# matrix size and data
N = 6
rand_matrix = -1 + 2*np.random.random((N, N))

# plotting matrix and colorbar
plt.imshow(rand_matrix, cmap='PiYG', vmin=-1, vmax=1)
plt.colorbar(label=r'$T/T_0$')

# writing value of each matrix element
for i in range(N):
    for j in range(N):
        c = round(rand_matrix[j,i], 2)
        plt.text(i, j, str(c), va='center', ha='center', fontsize=10)

# make index start from 1 instead of 0 in the plot
plt.xticks( np.arange(0, N), np.arange(0, N)+1 )
plt.yticks( np.arange(0, N), np.arange(0, N)+1 );
```

# Plotting bars

https://matplotlib.org/stable/gallery/lines_bars_and_markers/barchart.html#sphx-glr-gallery-lines-bars-and-markers-barchart-py

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
data1 = [20, 34, 30, 35, 27]
data2 = [25, 32, 34, 20, 25]
data3 = [21, 30, 10, 25, 22]
data4 = [28, 12, 14, 28, 22]
data5 = [32, 13, 24, 23, 21]

DATA   = [data1, data2, data3, data4, data5]
LABELS = ['data1', 'data2', 'data3', 'data4', 'data5']

Ndata = len(DATA) # 4 lists

x = np.arange(len(labels))  # the label locations
width_bar = 0.8  # the width of the bars

plt.figure()

for i_data in range(Ndata):
    disp = width_bar*(i_data/Ndata - 1/2)

    # when plotting more than one bar this strtucture makes all bars to be centered around the value x
    # it is important to use align='edge'
    plt.bar(x + disp,  DATA[i_data], width=width_bar/Ndata, label=LABELS[i_data], align='edge', bottom=0)

plt.bar(x,  DATA[0], width=width_bar, alpha=0.2)    

plt.xticks(x, labels)
plt.legend(ncol=Ndata, fontsize=10, loc='upper center')

plt.ylabel('N')


```

```python

# horizontal bars!

plt.figure()

for i_data in range(Ndata):
    disp = width_bar*(i_data/Ndata - 1/2)

    # when plotting more than one bar this strtucture makes all bars to be centered around the value x
    # it is important to use align='edge'
    # Now it changed from width to height!
    plt.barh(x + disp,  DATA[i_data], height=width_bar/Ndata, label=LABELS[i_data], align='edge')

plt.barh(x,  DATA[0], alpha=0.2)    

plt.yticks(x, labels)
plt.legend(ncol=1, fontsize=10, loc='upper right')

plt.xlabel('N')
```
